### Jogo da Memória Numérica

Este jogo desafia sua memória, pedindo para você memorizar sequências numéricas que aumentam de tamanho a cada acerto. Você terá um tempo limitado para digitar a sequência correta. Boa sorte!

```python
import random
import time
import sys

def generate_sequence(length):
    """Generates a random sequence of digits."""
    return ''.join([str(random.randint(0, 9)) for _ in range(length)])

def get_user_input_timed(sequence_length, time_limit):
    """Prompts the user for input with a time limit."""
    print(f"Você tem {time_limit} segundos para digitar a sequência.")
    start_time = time.time()

    # Use select for non-blocking input if available, or a simple input with a timer
    # For simplicity in Colab, we'll use a direct input and check time afterwards.
    # A more robust solution for timed input might involve threading or external libraries.
    print("Digite a sequência:")
    user_input = ''
    try:
        # This is a simplification; a true timed input is complex in standard Python input()
        # For Colab, the user types directly. We can't interrupt 'input()' once it starts.
        # So, we'll give the time limit as a guideline and check after input.
        user_input = input()
    except:
        pass # Handle potential EOF or other input errors

    end_time = time.time()
    if (end_time - start_time) > time_limit:
        print("Tempo esgotado!")
        return None
    return user_input

def play_game():
    score = 0
    sequence_length = 5
    time_limit = 5 # seconds
    max_rounds = 10

    print("Bem-vindo ao Jogo da Memória Numérica!")

    while score < max_rounds:
        print(f"\n--- Rodada {score + 1} ---")
        current_sequence = generate_sequence(sequence_length)

        print("Memorize esta sequência:")
        print(current_sequence)
        time.sleep(sequence_length + 1) # Display sequence for a bit longer than its length

        # Clear the console/output to hide the sequence for memory test
        # This is a hack for Colab/notebooks; actual console clearing is OS-dependent.
        # For a better UX, in a real application, you'd clear the screen.
        print("\n" * 50) # Scroll past the sequence
        print("Qual era a sequência?")

        user_attempt = get_user_input_timed(sequence_length, time_limit)

        if user_attempt is None:
            print("Você não conseguiu digitar a tempo. Game Over!")
            break

        if user_attempt == current_sequence:
            score += 1
            sequence_length += 1
            time_limit += 1
            print(f"Correto! Sua pontuação: {score}. Próxima sequência terá {sequence_length} dígitos. Tempo: {time_limit}s")
        else:
            print(f"Incorreto! A sequência correta era: {current_sequence}")
            print(f"Você errou. Game Over! Reiniciando o jogo.")
            score = 0
            sequence_length = 5
            time_limit = 5
            print("Vamos tentar de novo!")

    if score == max_rounds:
        print("\nParabéns você ganhou! Você completou todas as 10 rodadas!")

# Call the game function to start playing
play_game()
```
